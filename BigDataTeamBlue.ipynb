```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def heatmap(x, y, size):
    fig, ax = plt.subplots()
    
    # Mapping from column names to integer coordinates
    x_labels = [v for v in sorted(x.unique())]
    y_labels = [v for v in sorted(y.unique())]
    x_to_num = {p[1]:p[0] for p in enumerate(x_labels)} 
    y_to_num = {p[1]:p[0] for p in enumerate(y_labels)} 
    
    size_scale = 500
    ax.scatter(
        x=x.map(x_to_num), # Use mapping for x
        y=y.map(y_to_num), # Use mapping for y
        s=size * size_scale, # Vector of square sizes, proportional to size parameter
        marker='s' # Use square as scatterplot marker
    )
    
    # Show column labels on the axes
    ax.set_xticks([x_to_num[v] for v in x_labels])
    ax.set_xticklabels(x_labels, rotation=45, horizontalalignment='right')
    ax.set_yticks([y_to_num[v] for v in y_labels])
    ax.set_yticklabels(y_labels)
    
    
corr = sample_df.corr()
corr = pd.melt(corr.reset_index(), id_vars='index') # Unpivot the dataframe, so we can get pair of arrays for x and y
corr.columns = ['x', 'y', 'value']
heatmap(
    x=corr['x'],
    y=corr['y'],
    size=corr['value'].abs()
)
```

```python
# upsample N instance 
from sklearn.utils import resample

sample_df_majority = sample_df[sample_df.Delays=='Y']
sample_df_minority = sample_df[sample_df.Delays=='N']

sample_df_upsampled = resample(sample_df_minority, 
                                                    replace=True,     # sample with replacement
                                                    n_samples=12719,    # to match majority class
                                                    random_state=45) # reproducible results

sample_df = pd.concat([sample_df_majority, 
                                            sample_df_upsampled])
```

```python
# transform catogry into numeric value
x = sample_df.drop(['Delays','Year'],axis=1)
y = sample_df['Delays']

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

cat_transformer = Pipeline(steps=[("onehot", OneHotEncoder(handle_unknown='ignore'))])
cat_attr = ['Month','DayofMonth','DayOfWeek','UniqueCarrier','Origin','Dest']
preprocessor = ColumnTransformer(transformers=[("cat", cat_transformer, cat_attr)])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, KFold, cross_validate, cross_val_score
from sklearn.metrics import accuracy_score
import pickle


base_models = [("dt_clf", DecisionTreeClassifier(random_state=42)),
               ("rf_clf", RandomForestClassifier(n_estimators = 300,random_state=42)),
               ("svm_clf",svm.NuSVC(random_state=42,gamma='auto')),
               ('lr_clf', LogisticRegression(random_state=42))]

kfolds = 10
split = KFold(n_splits=kfolds, shuffle=True, random_state=42)

for name, model in base_models:
    # pack preprocessing of data and the model in a pipeline:
    model_steps = Pipeline(steps=[('preprocessor', preprocessor),
                                 ('model', model)])
    
    # get cross validation score for each model:
    cv_results = cross_val_score(model_steps, 
                                 x, y, 
                                 cv=split,
                                 scoring="accuracy",
                                 n_jobs=-1)
    # Save the model as a pickle in a file 
     
    
    # output:[]
    min_score = round(min(cv_results), 4)
    max_score = round(max(cv_results), 4)
    mean_score = round(np.mean(cv_results), 4)
    std_dev = round(np.std(cv_results), 4)
    print(f"{name} cross validation accuarcy score: {mean_score} +/- {std_dev} (std) min: {min_score}, max: {max_score}")
```

```python
#Random Forest 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, KFold, cross_validate, cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import joblib



x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.33, random_state = 1)

kfolds = 4
split = KFold(n_splits=kfolds, shuffle=True, random_state=4)

model_steps = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', RandomForestClassifier(n_estimators = 200,random_state=1))])


# get cross validation score for each model:
y_pred = model_steps.fit(x_train,y_train).predict(x_test)

joblib.dump(model_steps, 'rf_clf.pkl')
print('model saved')
print('\n')
matrix = confusion_matrix(y_test, y_pred)
print(matrix )
print('\n')
score = round(accuracy_score(y_test, y_pred),4)*100
print(f'Accuracy of ths model is {score}')
print('\n')
report = classification_report(y_test, y_pred)
print(report)
```
